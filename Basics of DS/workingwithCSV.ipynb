### usecols

```python
#student.csv -> name, roll, class, dept, cgpa
df = pd.read_csv('student.csv', usecols=['name', 'dept'])
# df only contains name and dept. columns
```

### skip rows and nrows

```python
df = pd.read_csv('file', skiprows=[0, 5])
#it skips 0 and 5 indexed rows
```

```python
df = pd.read_csv('file', nrows=100)
#loads first 100 rows
```

### encoding

```python
# UnicodeError can be fixed using encoding parameter
df = pd.read_csv('Zomato.csv', encoding='latin-1')
# find the encodig type in dataset
```

### bad_lines

```python
#file -> name;dept.
# but in some rows there are more than 2 columns, these are considers as bad lines
# error = raise an exception
# skip = skip bad lines without warning
# warn = raise warning and skip bad lines
df = pd.read_csv('file.csv', sep=';', on_bad_lines="warn")
```

### parse_dates

```python
# converts object to date-times
df = pd.read_csv('file', parse_dates=['date'])
```

### navalues

```python
# converts to NaN
df = pd.read_csv('file', na_values=['?', '-', 'm'])
```

### loading datasets into chunks

```python
df = pd.read_csv('file', chunksize=3)
#if file has 10 rows then each chunk has max of 3 rows
```
