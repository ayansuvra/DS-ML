# Bivariate and Multivariate Data Analysis

```python
import pandas as pd
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.head()
```

```python
import seaborn as sns
```

## Numeric-Numeric

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

```python
titanic = pd.read_csv('train.csv')
```

```python
titanic.head()
```

## Numeric-Categorical

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', hue='Sex')
```

```python
sns.boxplot(data=titanic, x='Sex', y='Age', hue='Survived')
```

```python
import matplotlib.pyplot as plt
```

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False) # alive
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False) #died
```

## Categorical - Categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
grouped = titanic.groupby('Pclass')['Survived'].mean()
```

```python
grouped * 100
```

```python
sns.barplot(data=titanic, x='Pclass', y='Fare', hue='Survived')
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

```python
sns.pairplot(titanic, hue='Embarked')
```
