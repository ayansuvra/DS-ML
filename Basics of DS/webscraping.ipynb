```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav'
headers = {"User-Agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'}
webpage = requests.get(url, headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'html.parser')
```

```python
soup.find_all('h1')[0].text.strip()
```

```python
soup.find_all('h2')
```

```python
for i in soup.find_all('h2'):
    print(i.text.strip())
```

```python
for i in soup.find_all('span', class_="companyCardWrapper__companyRatingCount"):
    print(i.text.strip())
```

```python
company = soup('div', class_="companyCardWrapper")
name = []
review_count = []
salary = []
jobs = []
for i in company:
    name.append(i.find('h2').text.strip())
    review_count.append(i.find('span', class_="companyCardWrapper__companyRatingCount").text.strip())
    salary.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[1].text)
    jobs.append(i.find_all('span', class_="companyCardWrapper__ActionCount")[3].text)
```

```python
import pandas as pd
```

```python
company_data = {
    'Name': name,
    'No. of ratings' : review_count,
    'salary' : salary,
    'jobs' : jobs
}
df = pd.DataFrame(company_data)
```
