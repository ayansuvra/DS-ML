```python
class MyAtm:
    # constructor
    def __init__(self):
        self.pin = None
        self.balance = 0
    # this function automatically gets activated when class is called
        print("Executed")
        self.menu()
    
    def menu(self):
        user_input = int(input("""
             1. create pin
             2. change pin
             3. check balance
             4. withdraw
             5. exit
         """))
        
        if user_input == 1:
            # create pin
            self.create_pin()
        elif user_input == 2:
            # change pin
            self.change_pin()
        elif user_input == 3:
            # check balance
            self.check_balance()
        elif user_input == 4:
            # withdraw
            self.withdraw()
        else:
            # exit
            pass
            
    def create_pin(self):
        user_pin = int(input('Enter Pin'))
        self.pin = user_pin

        user_balance = int(input('Enter Balance'))
        self.balance = user_balance

        print('PIN created succesfully')
        self.menu()

    def change_pin(self):
        old_pin = int(input('Enter Old PIN'))
        if old_pin == self.pin:
            new_pin = int(input('Enter New PIN'))
            self.pin = new_pin
            print('PIN updated successfully')
            self.menu()
        else:
            print('PIN not matched')
            self.menu()

    def check_balance(self):
        pin = int(input('Enter PIN'))
        if pin == self.pin:
            print('Balance : ',self.balance)
            self.menu()
        else:
            print('PIN not matched')
            self.menu()
        
    def withdraw(self):
        pin = int(input('Enter PIN'))
        if pin != self.pin:
            print('PIN not matched')
            self.menu()
        else:
            withdraw_amount = int(input('Enter Amount'))
            if withdraw_amount > self.balance:
                print('Not have enough money')
                print('Balance : ', self.balance)
            else:
                new_balance = self.balance - withdraw_amount
                self.balance = new_balance
                print('Balance : ', self.balance)
                self.menu()
```

## fraction

```python
class fraction:
    # parameterized constructor
    def __init__(apnadil, x, y):
        apnadil.num = x
        apnadil.den = y
        
    def __str__(apnadil): # triggers when object is being printed
        return '{}/{}'.format(apnadil.num, apnadil.den)

    def __add__(apnadil, uskadil): # triggers when addition is performmed between two objects
        new_num = apnadil.num*uskadil.den + apnadil.den*uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __sub__(apnadil, uskadil): # triggers when subtraction is performmed between two objects
        new_num = apnadil.num*uskadil.den - apnadil.den*uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __mul__(apnadil, uskadil): # triggers when multiplication is performmed between two objects
        new_num = apnadil.num * uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __truediv__(apnadil, uskadil): # triggers when division is performmed between two objects
        new_num = apnadil.num * uskadil.den
        new_den = apnadil.den * uskadil.num
        return '{}/{}'.format(new_num, new_den)
        
    def convert_to_decimal(self):
        return self.num / self.den
```

```python
fr = fraction(8, 2)
```

# 2D Co-ordinate System

```python
class Point:

    def __init__(self, x, y):
        self.x_cod = x
        self.y_cod = y

    def __str__(self): # coordination representation
        return '<{},{}>'.format(self.x_cod, self.y_cod)

    def euclidean_distance(self, other): # distance between two points
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5

    def origin_distance(self): # disinstance from origin
        return (self.x_cod**2 + self.y_cod**2)**0.5
        # return self.euclidean_distance(Point(0,0))
```

```python
class Line:

    def __init__(self, a, b, c):
        self.A = a
        self.B = b
        self.C = c

    def __str__(self): # line representation Ax + By + C = 0
        return '{}x + {}y + {} = 0'.format(self.A, self.B, self.C)

    def point_in_line(line, point):
        if line.A*point.x_cod + line.B*point.y_cod + line.C == 0:
            return "Point lies on Line"
        else:
            return "Point not lies on Line"

    def point_line_distance(line, point):
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 + line.B**2)**0.5

    def intersect_point(line1, line2):
        x = (line1.B*line2.C - line1.C*line2.B)/(line1.A*line2.B - line1.B*line2.A)
        y = (line1.C*line2.A - line1.A*line2.C)/(line1.A*line2.B - line1.B*line2.A)
        return print(Point(x, y))
```

# static Method

- created at clsss level
- accessed through class name
- object independent
- marked as static method

```python
class Person:
    __counter = 1
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        self.gender = gender
        self.id = Person.__counter
        Person.__counter += 1

    @staticmethod
    def get_id():
        return Person.__counter
```

```python
a1 = Person('Ayan', 19, 'M')
a2 = Person('Chayan', 18, 'M')
a3 = Person('Titli', 16, 'F')
```

```python
print(a1.id, a2.id, a3.id)
```

```python
print(Person.get_id())
```

```python
Person('a', 9, 'a')._Person__counter
```

# Aggregation

a class owns a class

```python
class Customer:

    def __init__(self, name, gender, address):
        self.name = name
        self.gender = gender
        self.address = address
    
    def print_address(self):
        print(self.address.get_city(), self.address.pin, self.address.state)

    def edit_profile(self, new_name, new_city, new_pin, new_state):
        self.name = new_name
        self.address.edit_address(new_city, new_pin, new_state)
        
class Address:

    def __init__(self, city, pin, state):
        self.__city = city
        self.pin = pin
        self.state = state
        
    def get_city(self):
        return self.__city

    def edit_address(self, new_city, new_pin, new_state):
        self.__city = new_city
        self.pin = new_pin
        self.state = new_state
```

```python
add1 = Address('Kolkata', 700001, 'West Bengal')
cust1 = Customer('Ayan', 'M', add1)
```

```python
cust1.print_address()
```

```python
cust1.edit_profile('ASB', 'Bardhaman', 712410, 'West Bengal')
```

```python
cust1.print_address()
```

### method overloading

In many programming languages like C++ or Java, you can define multiple methods with the same name but different parameter lists. This concept is called method overloading. Python does not support method overloading by default.
The code is more clean in reading.

```python
class Shape:
    def __init__(self):
        print('Area')
        
    def area(self, l, b): # executes when there are 2 arguments
        return l*b
        
    def area(self, r): # executes when there is only 1 arguments
        return 3.14*r*r
    
    # based on arguments the functions executes
```

```python
s = Shape()
```

```python
s.area(2, 3)
```

```python
# python does not support method overloadin, but it can be implemented by using default values
```

```python
class Shape:

    def __init__(self):
        print('Area')

    def area(self, a, b=0):
        if b==0:
            return 3.14*a*a
        else:
            return a*b
```

```python
s = Shape()
s.area(1)
```

```python
s.area(2, 4)
```

# Abstruction

```python
from abc import ABC, abstractmethod
# abstract base classes
class BankApp(ABC): # abstract class

    def __init__(self):
        pass
    def info(self):
        pass

    #abstract method
    @abstractmethod
    def security(self):
        pass
```

```python
class MobileApp(BankApp):

    def login(self):
        print('Logged In')
        
    def user(self):
        pass

    def security(self):
        pass
```

```python
a = BankApp() # You can't create a object of abstract class
```

```python
b = MobileApp()
```

a object cannot be created untill abstract method is applied on child class
