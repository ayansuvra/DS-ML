```python
class MyAtm:
    # constructor
    def __init__(self):
        self.pin = None
        self.balance = 0
    # this function automatically gets activated when class is called
        print("Executed")
        self.menu()
    
    def menu(self):
        user_input = int(input("""
             1. create pin
             2. change pin
             3. check balance
             4. withdraw
             5. exit
         """))
        
        if user_input == 1:
            # create pin
            self.create_pin()
        elif user_input == 2:
            # change pin
            self.change_pin()
        elif user_input == 3:
            # check balance
            self.check_balance()
        elif user_input == 4:
            # withdraw
            self.withdraw()
        else:
            # exit
            pass
            
    def create_pin(self):
        user_pin = int(input('Enter Pin'))
        self.pin = user_pin

        user_balance = int(input('Enter Balance'))
        self.balance = user_balance

        print('PIN created succesfully')
        self.menu()

    def change_pin(self):
        old_pin = int(input('Enter Old PIN'))
        if old_pin == self.pin:
            new_pin = int(input('Enter New PIN'))
            self.pin = new_pin
            print('PIN updated successfully')
            self.menu()
        else:
            print('PIN not matched')
            self.menu()

    def check_balance(self):
        pin = int(input('Enter PIN'))
        if pin == self.pin:
            print('Balance : ',self.balance)
            self.menu()
        else:
            print('PIN not matched')
            self.menu()
        
    def withdraw(self):
        pin = int(input('Enter PIN'))
        if pin != self.pin:
            print('PIN not matched')
            self.menu()
        else:
            withdraw_amount = int(input('Enter Amount'))
            if withdraw_amount > self.balance:
                print('Not have enough money')
                print('Balance : ', self.balance)
            else:
                new_balance = self.balance - withdraw_amount
                self.balance = new_balance
                print('Balance : ', self.balance)
                self.menu()
```

## fraction

```python
class fraction:
    # parameterized constructor
    def __init__(apnadil, x, y):
        apnadil.num = x
        apnadil.den = y
        
    def __str__(apnadil): # triggers when object is being printed
        return '{}/{}'.format(apnadil.num, apnadil.den)

    def __add__(apnadil, uskadil): # triggers when addition is performmed between two objects
        new_num = apnadil.num*uskadil.den + apnadil.den*uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __sub__(apnadil, uskadil): # triggers when subtraction is performmed between two objects
        new_num = apnadil.num*uskadil.den - apnadil.den*uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __mul__(apnadil, uskadil): # triggers when multiplication is performmed between two objects
        new_num = apnadil.num * uskadil.num
        new_den = apnadil.den * uskadil.den
        return '{}/{}'.format(new_num, new_den)

    def __truediv__(apnadil, uskadil): # triggers when division is performmed between two objects
        new_num = apnadil.num * uskadil.den
        new_den = apnadil.den * uskadil.num
        return '{}/{}'.format(new_num, new_den)
        
    def convert_to_decimal(self):
        return self.num / self.den
```

```python
fr = fraction(8, 2)
```
