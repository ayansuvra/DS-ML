### Dynamic Typing and Dynamic Binding

no need to specify data types before creating a variable,
variables in python can store value with any datatype even after intialising with different data types

```python
a, b, c = 1, 3, 7
print(a, b, c)
```

```python
a = b = c = 69
print(a, b, c)
```

```python
# name of a varible can be _, but not a reserve keyword
_ = 'name'
print(_)
```

```python
if = 'not valid'
print(if)
```

## Operators

```python
print(5 + 2)
print(5 - 2)
print(5 * 2)
print(5 / 2)
print(5 // 2)
print(5 ** 2)
```

```python
print(1 and 0)
print(1 or 0)
print(not 1)
```

### if-else

```python
email = input("Enter Email : ")
password = input("Enter Password : ")

if email == "asb@gmail.com" and password == "12345":
    print('Welcome')
else:
    print('Wrong inputs')
```

```python
# min of 3 numbers

a = int(input('1st No : '))
b = int(input('2nd No : '))
c = int(input('3rd No : '))

if b<a and b<c:
    print(b, 'is smallest')
elif a<c and a<b:
    print(a, 'is smallest')
else:
    print(c, 'is smallest')
```
