## String Functions

```python
len("Hello, everyone") # length of string
```

```python
max("hello, everyone") # maximum value present in string
```

```python
min("hello, everyone") # minimum value present in string
```

```python
sorted("hello, everyone", reverse=True) # sort
```

this functions can also be applied list, touple and sets.

```python
# capitalize
s = "hello, world"
s.capitalize()
```

```python
# title
s.title()
```

```python
# upper
s.upper()
```

```python
# lower
s.lower()
```

```python
# swapcase - lower to upper and upper to lower
"HeLLo wOrLD".swapcase()
```

```python
# count - to get the frequency of a substring
"Hi, I am ASB".count('i')
```

```python
# find - to get the index, returns -1 is the substring is not present in the string
"Hi, I am ASB".find('i')
```

```python
# index - same as find, but it throws an error if the substring is not present in string
"Hi, I am ASB".index('i')
```

```python
M = "Hello everyone"
M.find('x')
```

```python
M.index('X')
```

```python
# startwith and endwith
M.startswith('H')
```

```python
M.startswith("y")
```

```python
M.endswith('e')
```

```python
M.endswith('.')
```

```python
# format
name = "ASB"
gender = "Male"

"Hi, my name is {}. My gender is {}".format(name, gender)
```

```python
"Hi, my name is {1}. My gender is {0}".format(gender, name)
```

```python
# isalnum() - checks if the string only contains alpha or numeric values
"Hey1343".isalnum()
```

```python
"Hey^&23".isalnum()
```

```python
# isalpha() - checks if the string only contains alpha values
"hey".isalpha()
```

```python
"hey23".isalpha()
```

```python
# isdigit() - checks for digit
'134'.isdigit()
```

```python
# isidentifier() - checks if the identifier is valid or not
"_name".isidentifier()
```

```python
"first-name".isidentifier()
```

```python
'_'.isidentifier()
```

```python
'+'.isidentifier()
```

### split-join

```python
"Hey, my name is Ayan".split() # by default it splits the string by spaces
```

```python
"Hi, my name is Ayan".split('i')
```

```python
# join
" ".join(['Hey,', 'my', 'name', 'is', 'Ayan'])
```

```python
'-'.join(['Hey,', 'my', 'name', 'is', 'Ayan'])
```

```python
'i'.join(['H', ', my name ', 's Ayan'])
```

```python
"Hi, my name is Ayan".replace('Ayan', 'Suvra') #replace substrings
```

```python
# strip - remove unwanted spaces from first and last
"Hey, my name is Ayan".strip()
```

```python
"             Ayan                                Suvra                 Bosu              ".strip()
```
