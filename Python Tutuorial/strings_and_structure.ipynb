## List comprehension

l = [expression for item in iterable if condition == True]

```python
M = [i for i in range(1, 11)]
M
```

```python
j = 3
[j*i for i in M]
```

```python
# make a square list
n = [1, 2, 3, 4, 5]
n_square = [i*i for i in n]
```

```python
n_square
```

```python
# print all numbers which are divisible by 50 in the range of 1 to 50
div_list = [i for i in range(1, 51) if i%5 == 0]
```

```python
div_list
```

```python
# nested list comprehension
[[i*j for i in range(1, 4)] for j in range(1, 4)]
```

```python
l1 = [1, 2, 3, 4]
l2 = [5, 6, 7, 8]
l = [i*j for i in l1 for j in l2]
```

```python
l
```

```python
# zip function
list(zip(l1, l2))
```

```python
[i+j for i,j in zip(l1, l2)]
```

```python
[i+j for i in l1 for j in l2]
```
